```python
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer(min_df = 0, ngram_range = (1, 2))
genre_mst = vect.fit_transform(movies_df['genres_literal'])
```

```python
genre_mat = genre_mst.toarray()
```

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=0):
    title_movie = df[df['title'] == title_name] # 해당 영화 데이터 row
    title_index = title_movie.index.values # 해당 영화의 index
     
    similarity_indexes = sorted_ind[title_index, :(top_n)] 
    # sorted_ind matrix : row - 영화인덱스 순서, columns - 유사도 순으로 영화 인덱스가 올림차순 정렬
    # sorted_ind: 유사도 matrix(해당영황 row index, 왼쪽에서 n개 칼럼)
    similarity_indexes = similarity_indexes.reshape(-1)
    record = df.iloc[similarity_indexes]
    
    return record
```

```python
df = movies_df
sorted_ind = genre_sim_sorted_ind
title_name = 'The Godfather'
top_n = 5

find_sim_movie(df, sorted_ind, title_name, top_n)
```

```python
def weighted_vote_average(record):
    v = record['vote_count']
    r = record['vote_average']
    
    record = record[(v > record['vote_count'].mean()) & (r > record['vote_average'].mean())]
    record = record.popularity.sort_values(ascending=False)
    
    return record
```

```python
def find_sim_movie_popular(df, sorted_ind, title_name, top_n=0):
    title_movie = df[df['title'] == title_name] # 해당 영화 데이터 row
    title_index = title_movie.index.values # 해당 영화의 index
     
    similarity_indexes = sorted_ind[title_index, :(top_n)] 
    # sorted_ind matrix : row - 영화인덱스 순서, columns - 유사도 순으로 영화 인덱스가 올림차순 정렬
    # sorted_ind: 유사도 matrix(해당영황 row index, 왼쪽에서 n개 칼럼)
    similarity_indexes = similarity_indexes.reshape(-1)
   
    record = df.iloc[similarity_indexes]

    final_df = weighted_vote_average(record)
    
    return final_df
```

```python
df = movies_df
sorted_ind = genre_sim_sorted_ind
title_name = 'The Godfather'
top_n = 5

df.loc[find_sim_movie_popular(df, sorted_ind, title_name, top_n = 10).index]
```
